```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import sklearn
from sklearn.utils import shuffle

import math
from sklearn import svm
import sklearn.metrics as metric

# from sklearn.model_selection import train_test_split

%matplotlib inline

df = pd.read_csv("C:/Users/shahh/OneDrive/Desktop/Research Paper - KS/Final CSV/April/final_data.csv")
# df = df.drop(columns=['File Name','Number of cells','Cell Number'])
df = shuffle(df)
df

```

```python
df = df.drop(columns=['File Name'])
df
```

```python
X = df.drop('IsShade', axis=1)
y = df['IsShade']
```

```python

from sklearn.model_selection import train_test_split
# implementing train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.34, random_state=44)

```

```python
svc = svm.SVC(kernel='linear').fit(X_train,y_train)
svc
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel='linear')
svm_classifier.fit(X_train,y_train)
svm_classifier_predict = svm_classifier.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

rfc_cv_score = cross_val_score(svm_classifier, X, y, cv=10, scoring='roc_auc')
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, svm_classifier_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, svm_classifier_predict))
print('\n')
print("=== All AUC Scores ===")
# print(rfc_cv_score)
# print('\n')
# print("=== Mean AUC Score ===")
# print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```

```python
from sklearn.metrics import accuracy_score

print("Train Accuracy :: ", accuracy_score(y_train, svm_classifier.predict(X_train)))
print("Test Accuracy  :: ", accuracy_score(y_test, svm_classifier_predict))
```
